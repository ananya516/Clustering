1. What is Unsupervised Learning?

Unsupervised learning is a machine learning approach where the model identifies patterns from input data without using labeled outputs. It includes tasks like clustering and dimensionality reduction.

2. How does K-Means Clustering Algorithm Work?

Randomly initialize k centroids.

Assign each point to the nearest centroid.

Recalculate centroids by averaging assigned points.

Repeat until centroids no longer change significantly.

3. What is a Dendrogram in Hierarchical Clustering?

A dendrogram is a tree diagram that shows the merging or splitting of clusters step-by-step in hierarchical clustering. It helps in selecting the number of clusters visually.

4. What is the Main Difference between K-Means and Hierarchical Clustering?

K-Means: Flat clustering, needs predefined k.

Hierarchical: Tree-structured clustering, does not require k upfront.

5. What are the Advantages of DBSCAN over K-Means?

Detects arbitrarily shaped clusters.

Handles noise points.

No need to pre-specify the number of clusters.

6. When Would You Use Silhouette Score in Clustering?

To evaluate clustering quality by measuring how similar a sample is to its own cluster vs. other clusters. Useful for choosing the optimal number of clusters.

7. What are the Limitations of Hierarchical Clustering?

Computationally expensive (not scalable to large datasets).

Sensitive to outliers and noise.

Merging decisions are irreversible.

8. Why is Feature Scaling Important in K-Means?

K-Means uses Euclidean distance, which is sensitive to the scale of features. Unscaled data can bias clustering toward features with larger numerical ranges.

9. How Does DBSCAN Identify Noise Points?

Points that do not belong to any dense region (i.e., not a core point or reachable from one) are labeled as noise.

10. Define Inertia in K-Means.

Inertia is the sum of squared distances between each point and its assigned centroid. Lower inertia means tighter clusters.

11. What is the Elbow Method in K-Means Clustering?

A method to determine the optimal number of clusters k by plotting inertia vs. k and identifying the "elbow" point where the inertia reduction slows.

12. Describe the Concept of “Density” in DBSCAN.

Density is defined by the number of points within a given radius (eps). A region is dense if it has at least min_samples within eps.

13. Can Hierarchical Clustering Be Used on Categorical Data?

Yes, but requires a suitable distance metric (e.g., Hamming distance) or encoding like one-hot encoding.

14. What Does a Negative Silhouette Score Indicate?
The sample may be assigned to the wrong cluster since it’s closer to points in another cluster than its own.

15. Explain “Linkage Criteria” in Hierarchical Clustering.

Determines how distances between clusters are calculated:

Single: min distance

Complete: max distance

Average: average distance

Ward: minimizes variance

16. Why Might K-Means Perform Poorly on Data with Varying Cluster Sizes or Densities?

K-Means assumes equal-sized, spherical clusters. It struggles with uneven densities, sizes, or non-convex shapes.

17. What are the Core Parameters in DBSCAN?

eps: Radius for neighborhood

min_samples: Minimum number of points to form a dense region
These determine core points and clustering behavior.

18. How Does K-Means++ Improve K-Means Initialization?

K-Means++ chooses initial centroids more carefully, spreading them out to avoid poor convergence and improve accuracy.

19. What is Agglomerative Clustering?

A bottom-up hierarchical clustering method where each data point starts as its own cluster, and clusters are merged iteratively.

20. What Makes Silhouette Score Better than Just Inertia for Evaluation?

Inertia only decreases with more clusters, but Silhouette Score balances compactness and separation, giving a more reliable evaluation of clustering quality.



```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

X, y = make_blobs(n_samples=300, centers=4, random_state=42)

kmeans = KMeans(n_clusters=4, random_state=42)
labels = kmeans.fit_predict(X)

plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=200, c='red', marker='X')
plt.title("K-Means Clustering with 4 Centers")
plt.show()
```

```python
from sklearn.datasets import make_moons
from sklearn.cluster import DBSCAN
import numpy as np

X, _ = make_moons(n_samples=300, noise=0.1, random_state=42)
dbscan = DBSCAN(eps=0.2, min_samples=5)
labels = dbscan.fit_predict(X)

# Outliers are labeled as -1
outliers = X[labels == -1]
core = X[labels != -1]

plt.scatter(core[:, 0], core[:, 1], c=labels[labels != -1], cmap='viridis')
plt.scatter(outliers[:, 0], outliers[:, 1], color='red', label='Outliers')
plt.title("DBSCAN on Moons Dataset")
plt.legend()
plt.show()
```

```python
from sklearn.datasets import load_digits
from sklearn.decomposition import PCA

digits = load_digits()
X = PCA(n_components=2).fit_transform(digits.data)

kmeans = KMeans(n_clusters=10, random_state=42)
labels = kmeans.fit_predict(X)

plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='tab10')
plt.title("KMeans Clustering on Digits (PCA 2D)")
plt.show()
```

```python
from sklearn.metrics import silhouette_score

X, _ = make_blobs(n_samples=300, centers=4, random_state=42)
scores = []
for k in range(2, 6):
    km = KMeans(n_clusters=k, random_state=42).fit(X)
    score = silhouette_score(X, km.labels_)
    scores.append(score)

plt.bar(range(2, 6), scores)
plt.xlabel("Number of clusters")
plt.ylabel("Silhouette Score")
plt.title("Silhouette Scores for Different K")
plt.show()
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage

Z = linkage(X, method='average')
plt.figure(figsize=(10, 5))
dendrogram(Z)
plt.title("Hierarchical Clustering Dendrogram (Average Linkage)")
plt.show()
```

```python
from matplotlib.colors import ListedColormap
from sklearn.datasets import make_blobs

X, _ = make_blobs(n_samples=300, centers=3, cluster_std=2.0, random_state=42)
kmeans = KMeans(n_clusters=3, random_state=42).fit(X)

h = 0.02
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

Z = kmeans.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.contourf(xx, yy, Z, alpha=0.3)
plt.scatter(X[:, 0], X[:, 1], c=kmeans.labels_, cmap='viridis')
plt.title("KMeans with Decision Boundaries")
plt.show()
```

```python
X, _ = make_blobs(n_samples=300, centers=3, random_state=42)
agg = AgglomerativeClustering(n_clusters=3, linkage='complete')
labels = agg.fit_predict(X)

plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis')
plt.title("Agglomerative Clustering (Complete Linkage)")
plt.show()
```

```python
inertias = []
for k in range(2, 7):
    km = KMeans(n_clusters=k, random_state=42).fit(X)
    inertias.append(km.inertia_)

plt.plot(range(2, 7), inertias, marker='o')
plt.title("Elbow Method on Breast Cancer Data")
plt.xlabel("K")
plt.ylabel("Inertia")
plt.show()
```
